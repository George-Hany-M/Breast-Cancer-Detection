# <font color='DarkSlateGrey'>⚑ Splitting the data and target</font>

```python
# X Data
X = data.drop(['diagnosis'], axis=1)
print('X Data is : \n' , X.head())
print()
print('X shape is : ' , X.shape)
print('-' * 100)
print()

# y Data
y = data['diagnosis']
print('y Data is : \n' , y.head())
print('y shape is : ' , y.shape)

#X = X.to_numpy()
#y = y.to_numpy()
#y = y.reshape(-1, 1)
```

# <font color='DarkSlateGrey'>⚑ Data Partitioning / Train Test Split</font>

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.325, random_state=42, shuffle=True)

# Splitted Data
print('X_train shape is ' , X_train.shape)
print('X_test shape is ' , X_test.shape)
print('y_train shape is ' , y_train.shape)
print('y_test shape is ' , y_test.shape)
```

# <font color='DarkSlateGrey'>⚑ Data Scaling / Feature Scaling</font>

## <font color='SaddleBrown'>- Scaling guarantees that all features are on a comparable scale and have comparable ranges.</font>
## <font color='SaddleBrown'>- Algorithm performance improvement.</font>

```python
# Standardization (Z-Score Normalization) 
# StandardScaler for Data

scaler = StandardScaler()

# Fit the scaler on the training data
scaler.fit(X_train)

# Transform the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test) # Generalizing what you learned in "Training Data" ...

# "fit" => This step calculates the mean and standard deviation of each feature in the training data.
# It's important to fit the scaler only on the training data, not the test data.
```

# <font color='DarkSlateGrey'>⚑ Applying "Binary Logistic Regression Model"</font>

```python
Model = LogisticRegression()
Model.fit(X_train_scaled, y_train)

# Calculating Prediction
y_pred = Model.predict(X_test_scaled)
```

```python
y_predection = pd.DataFrame(data={'Predicted Diagnosis': y_pred.flatten()})
print(y_predection)

# The "flatten()" function in Python is used to convert a multi-dimensional array or matrix into .....
# a one-dimensional array
```

```python
print('Actual Diagnosis')
y_test
```

# <font color='DarkSlateGrey'>⚑ Evaluation of the model</font>

```python
CM = confusion_matrix(y_true=y_test, y_pred=y_pred)
ConfusionMatrixDisplay(CM, display_labels=data['diagnosis'].unique()).plot()
plt.title('Confusion Matrix Without Normalization')
plt.show()
```

```python
CM = confusion_matrix(y_true=y_test, y_pred=y_pred, normalize='true')
ConfusionMatrixDisplay(CM, display_labels=data['diagnosis'].unique()).plot()
plt.title('Normalized Confusion Matrix \n (Values between 0 and 1)')
plt.show()
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
Accuracy = accuracy_score(y_test, y_pred)
print(f'➤➤➤ Accuracy Score : {Accuracy * 100 : .2f} %\n')
# Accuracy = (TP + TN) / (TP + TN + FP + FN)

Precision = precision_score(y_test, y_pred)
print(f'➤➤➤ Precision Score : {Precision * 100 : .2f} %\n')
# Precision = TP / (TP + FP)

Recall = recall_score(y_test, y_pred)
print(f'➤➤➤ Recall Score : {Recall * 100 : .2f} %\n')
# Recall = TP / (TP + FN)
```

```python
Scores = [Accuracy, Precision, Recall]
Score_Names = ['Accuracy', 'Precision', 'Recall']

# Plot
plt.figure(figsize=(7, 5))
plt.pie(Scores, labels=Score_Names, autopct='%1.2f%%', startangle=140)
plt.axis('equal')
plt.show()
```
